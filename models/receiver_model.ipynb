```python
college_summary = receiverDATA.groupby('school')[['recYDS', 'nfl_recYDS', 'nfl_YPG', 'YPG', 'RPG', 'nfl_RPG', 'TDPG', 'nfl_TDPG']].mean().reset_index()

sns.regplot(x='recYDS', y='nfl_recYDS', data=college_summary)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='recYDS',
    y='nfl_recYDS',
    hue='school',
    data=college_summary,
    palette='tab20'
)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
selected_schools = ['Ohio St.', 'Florida', 'Georgia', 'LSU', 'Alabama', 'USC', 'North Carolina', "Michigan"]
filtered_df = college_summary[college_summary['school'].isin(selected_schools)]
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='recYDS',
    y='nfl_YPG',
    hue='school',
    data=filtered_df,
    s=100
)
plt.title('Average College vs NFL Receiving Yards (Selected Schools)')
plt.show()
```
